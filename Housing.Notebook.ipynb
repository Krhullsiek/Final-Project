```python
# Function to calculate correlation coefficient between two arrays
def corr(x, y, **kwargs):
    
    # Calculate the value
    coef = np.corrcoef(x, y)[0][1]
    # Make the label
    label = r'$\rho$ = ' + str(round(coef, 2))
    
    # Add the label to the plot
    ax = plt.gca()
    ax.annotate(label, xy = (0.2, 0.95), size = 25, xycoords = ax.transAxes)


# Map the plots to the locations
g = sn.PairGrid(df2, palette=["red"])
g.map_upper(plt.scatter, s=10)
g.map_upper(corr)
g.map_diag(sn.displot, kde=True)
g.map_lower(sn.kdeplot, cmap="Blues_d")
plt.savefig("PearsonRGrid.png")
```

```python
X=corrDf.drop("price", axis=1)
y=corrDf['price']

print(X.shape, y.shape)

```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=101, test_size=0.3)
scaler = MinMaxScaler().fit(X_train)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print('Max: ',X_train.max())
print('Min: ', X_train.min())
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))


model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mse')

model.fit(x=X_train,y=y_train.values,
          validation_data=(X_test,y_test.values),
          batch_size=128,epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)

plt.figure(figsize=(15,5))
sn.lineplot(data=losses,lw=3)
plt.xlabel('Epochs')
plt.ylabel('')
plt.title('Training Loss per Epoch')
sn.despine()
```

```python
predictions = model.predict(X_test)
```

```python
f, axes = plt.subplots(1, 2,figsize=(15,5))

# Our model predictions
plt.scatter(y_test,predictions)

# Perfect predictions
plt.plot(y_test,y_test,'r')

errors = y_test.values.reshape(6484, 1) - predictions
sn.histplot(errors, ax=axes[0])

sn.despine(left=True, bottom=True)
axes[0].set(xlabel='Error', ylabel='', title='Error Histogram')
axes[1].set(xlabel='Test True Y', ylabel='Model Predictions', title='Model Predictions vs Perfect Fit')
```
