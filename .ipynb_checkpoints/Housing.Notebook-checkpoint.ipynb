```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np

from numpy.random import seed
seed(1)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('kc_house_data.csv')
```

```python
df.head()
```

```python
# df['date'] = pd.to_datetime(df["date"])
# df.head()

df['date'] = pd.to_datetime(df['date'])

df['month'] = df['date'].apply(lambda date:date.month)
df['year'] = df['date'].apply(lambda date:date.year)

df = df.drop('date',axis=1)

df.columns.values
df.head()
```

```python
list(df.columns)
```

```python
df.isnull().sum()
```

```python
corrDf = df.drop(columns=['id', 'zipcode'])
corrMatrix = corrDf.corr()
print (corrMatrix)
```

```python
sn.heatmap(corrMatrix, annot=True, cmap="YlGn")
plt.rcParams["figure.figsize"] = (20,20)
plt.savefig('housing.png')
plt.show()
```

```python
corrDf.dtypes
```

```python
X=corrDf.drop("price", axis=1)
y=corrDf['price']

print(X.shape, y.shape)

```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=101, test_size=0.3)
scaler = MinMaxScaler().fit(X_train)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print('Max: ',X_train.max())
print('Min: ', X_train.min())
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))


model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mse')

model.fit(x=X_train,y=y_train.values,
          validation_data=(X_test,y_test.values),
          batch_size=128,epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)

plt.figure(figsize=(15,5))
sn.lineplot(data=losses,lw=3)
plt.xlabel('Epochs')
plt.ylabel('')
plt.title('Training Loss per Epoch')
sn.despine()
```

```python
predictions = model.predict(X_test)
```

```python
f, axes = plt.subplots(1, 2,figsize=(15,5))

# Our model predictions
plt.scatter(y_test,predictions)

# Perfect predictions
plt.plot(y_test,y_test,'r')

errors = y_test.values.reshape(6484, 1) - predictions
sn.histplot(errors, ax=axes[0])

sn.despine(left=True, bottom=True)
axes[0].set(xlabel='Error', ylabel='', title='Error Histogram')
axes[1].set(xlabel='Test True Y', ylabel='Model Predictions', title='Model Predictions vs Perfect Fit')
```

```python
single_house = df.drop('price',axis=1).iloc[0]
print(f'Features of new house:\n{single_house}')

# reshape the numpy array and scale the features
single_house = scaler.transform(single_house.values.reshape(-1, 21))

# run the model and get the price prediction
print('\nPrediction Price:',model.predict(single_house)[0,0])

# original price
print('\nOriginal Price:', df.iloc[0]['price'])
```
